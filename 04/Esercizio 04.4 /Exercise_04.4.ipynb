## Argon and Krypton

Come al solito è necessario tenere conto dell'equilibrazione, di conseguenza è stata aumentata la statistica su cui fare l'analisi dei dati.

```python
import numpy as np
import math
import matplotlib 
import matplotlib.pyplot as plt

N=500
sum_prog=np.zeros(N)
err_prog=np.zeros(N)
x=np.arange(N)

#Data about Argon
sigma1=0.34*10**(-9)
rap1=120
mass1=39.948*1.66054*10**(-27)
K=1.3806*10**(-23)
eps1=rap1*K

#Data about Krypton
sigma2=0.364*10**(-9)
rap2=164
mass2=83.798*1.66054*10**(-27)
eps2=rap2*K

for i in range(3):
    
    if(i==0):
        print("GAS PHASE")
        file1=open("gas/ave_etot.gas")
        file2=open("gas/ave_ekin.gas")
        file3=open("gas/ave_epot.gas")
        file4=open("gas/ave_temp.gas")
        file5=open("gas/ave_press.gas")
    if(i==1):
        print("LIQUID PHASE")
        file1=open("liquid/ave_etot.liquid")
        file2=open("liquid/ave_ekin.liquid")
        file3=open("liquid/ave_epot.liquid")
        file4=open("liquid/ave_temp.liquid")
        file5=open("liquid/ave_press.liquid")
    if(i==2):
        print("SOLID PHASE")
        file1=open("solid/ave_etot.solid")
        file2=open("solid/ave_ekin.solid")
        file3=open("solid/ave_epot.solid")
        file4=open("solid/ave_temp.solid")
        file5=open("solid/ave_press.solid")
    
    #Total Energy
    
    line1=file1.readlines()
    for i in range(N):
        a=line1[i].split(" ")
        sum_prog[i]=a[0]
        err_prog[i]=a[1]
    file1.close()
    
    plt.figure()
        #Argon
    plt.subplot(121)
    plt.errorbar(x,sum_prog*eps1,yerr=err_prog*eps1)
    plt.xlabel("steps")
    plt.ylabel("$E_{tot}$/N (J)")
    plt.grid(True)
    plt.title("Argon: $E_{tot}$/N")
    
        #Krypton
    plt.subplot(122)
    plt.errorbar(x,sum_prog*eps2,yerr=err_prog*eps2, color='r')
    plt.xlabel("steps")
    plt.ylabel("$E_{tot}$/N (J)")
    plt.grid(True)
    plt.title("Krypton: $E_{tot}$/N")
    
    plt.subplots_adjust(bottom=0.1, right=2, top=0.9)
    plt.show()
    
    #Kinetic Energy
    
    line2=file2.readlines()
    for i in range(N):
        a=line2[i].split(" ")
        sum_prog[i]=a[0]
        err_prog[i]=a[1]
    file2.close()
    
    plt.figure()
        #Argon
    plt.subplot(121)
    plt.errorbar(x,sum_prog*eps1,yerr=err_prog*eps1)
    plt.xlabel("steps")
    plt.ylabel("$E_{kin}$/N (J)")
    plt.grid(True)
    plt.title("Argon: $E_{kin}$/N")
    
        #Krypton
    plt.subplot(122)
    plt.errorbar(x,sum_prog*eps2,yerr=err_prog*eps2,color='r')
    plt.xlabel("steps")
    plt.ylabel("$E_{kin}$/N (J)")
    plt.grid(True)
    plt.title("Krypton: $E_{kin}$/N")
    
    plt.subplots_adjust(bottom=0.1, right=2, top=0.9)
    plt.show()

    #Potential Energy
    
    line3=file3.readlines()
    for i in range(N):
        a=line3[i].split(" ")
        sum_prog[i]=a[0]
        err_prog[i]=a[1]
    file3.close()
    
    plt.figure()
        #Argon
    plt.subplot(121)
    plt.errorbar(x,sum_prog*eps1,yerr=err_prog*eps1)
    plt.xlabel("steps")
    plt.ylabel("$E_{pot}$/N (J)")
    plt.grid(True)
    plt.title("Argon: $E_{pot}$/N")
    
        #Krypton
    plt.subplot(122)
    plt.errorbar(x,sum_prog*eps2,yerr=err_prog*eps2, color='r')
    plt.xlabel("steps")
    plt.ylabel("$E_{pot}$/N (J)")
    plt.grid(True)
    plt.title("Krypton: $E_{pot}$/N")
    
    plt.subplots_adjust(bottom=0.1, right=2, top=0.9)
    plt.show()

    #Temperature
    
    line4=file4.readlines()
    for i in range(N):
        a=line4[i].split(" ")
        sum_prog[i]=a[0]
        err_prog[i]=a[1]
    file4.close()
    
    plt.figure()
        #Argon
    plt.subplot(121)
    plt.errorbar(x,sum_prog*rap1,yerr=err_prog*rap1)
    plt.xlabel("steps")
    plt.ylabel("T (K)")
    plt.grid(True)
    plt.title("Argon: Temperature")
    
        #Krypton
    plt.subplot(122)
    plt.errorbar(x,sum_prog*rap2,yerr=err_prog*rap2, color='r')
    plt.xlabel("steps")
    plt.ylabel("T (K)")
    plt.grid(True)
    plt.title("Krypton: Temperature")
    
    plt.subplots_adjust(bottom=0.1, right=2, top=0.9)
    plt.show()

    #Pressure
    
    line5=file5.readlines()
    for i in range(N):
        a=line5[i].split(" ")
        sum_prog[i]=a[0]
        err_prog[i]=a[1]
    file5.close()
    
    plt.figure()
        #Argon
    plt.subplot(121)
    plt.errorbar(x,sum_prog*eps1/(sigma1**3),yerr=err_prog*eps1/(sigma1**3))
    plt.xlabel("steps")
    plt.ylabel("P (Pa)")
    plt.grid(True)
    plt.title("Argon: Pressure")
    
        #Krypton
    plt.subplot(122)
    plt.errorbar(x,sum_prog*eps2/(sigma2**3),yerr=err_prog*eps2/(sigma2**3), color='r')
    plt.xlabel("steps")
    plt.ylabel("P (Pa)")
    plt.grid(True)
    plt.title("Krypton: Pressure")
    
    plt.subplots_adjust(bottom=0.1, right=2, top=0.9)
    plt.show()
```
