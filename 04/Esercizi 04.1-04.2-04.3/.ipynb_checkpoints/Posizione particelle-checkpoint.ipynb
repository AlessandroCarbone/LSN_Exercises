```python
import numpy as np
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

N=108
x=np.zeros(N)
y=np.zeros(N)
z=np.zeros(N)

x_old=np.zeros(N)
y_old=np.zeros(N)
z_old=np.zeros(N)

file=open("config.0")
line=file.readlines()
for i in range(N):
    data=line[i].split()
    x[i]=data[0]
    y[i]=data[1]
    z[i]=data[2]
file.close()

file=open("old.final")
line=file.readlines()
for i in range(N):
    data=line[i].split()
    x_old[i]=data[0]
    y_old[i]=data[1]
    z_old[i]=data[2]
file.close()

ax = Axes3D(plt.figure())
ax.scatter(x,y,z, color='r')
ax.scatter(x_old,y_old,z_old, color='b')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title("Posizione particelle")
ax.view_init(35, 35)
```
