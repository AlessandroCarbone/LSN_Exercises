### Osservazioni


- Come al solito è necessario tenere conto dell'equilibrazione, tuttavia la simulazione è stata condotta con i dati relativi ad un gas. Tale stato si distingue dagli altri per la difficoltà nell'equilibrazione, come è possibile osservare nei grafici che seguono.


- I risultati sono ottenuti leggendo la configurazione iniziale dal quella finale della simulazione precedente e sono espressi in unità naturali.

```python
import numpy as np
import math
import matplotlib 
import matplotlib.pyplot as plt

N=200
sum_prog=np.zeros(N)
err_prog=np.zeros(N)
x=np.arange(N)

#Total Energy
file1=open("Results/ave_etot.dat")
line1=file1.readlines()
for i in range(N):
    a=line1[i].split(" ")
    sum_prog[i]=a[0]
    err_prog[i]=a[1]
file1.close()

plt.errorbar(x,sum_prog,yerr=err_prog)
plt.xlabel("steps")
plt.ylabel("$E_{tot}$/N")
plt.grid(True)
plt.title("(Exercise 04.2): $E_{tot}$/N")
plt.show()

#Kinetic Energy
file2=open("Results/ave_ekin.dat")
line2=file2.readlines()
for i in range(N):
    a=line2[i].split(" ")
    sum_prog[i]=a[0]
    err_prog[i]=a[1]
file2.close()
    
plt.figure()
plt.errorbar(x,sum_prog,yerr=err_prog)
plt.xlabel("steps")
plt.ylabel("$E_{kin}$/N")
plt.grid(True)
plt.title("(Exercise 04.2): $E_{kin}$/N")
plt.show()

#Potential Energy
file3=open("Results/ave_epot.dat")
line3=file3.readlines()
for i in range(N):
    a=line3[i].split(" ")
    sum_prog[i]=a[0]
    err_prog[i]=a[1]
file3.close()
    
plt.figure()
plt.errorbar(x,sum_prog,yerr=err_prog)
plt.xlabel("steps")
plt.ylabel("$E_{pot}$/N")
plt.grid(True)
plt.title("(Exercise 04.2): $E_{pot}$/N")
plt.show()

#Temperature
file4=open("Results/ave_temp.dat")
line4=file4.readlines()
for i in range(N):
    a=line4[i].split(" ")
    sum_prog[i]=a[0]
    err_prog[i]=a[1]
file4.close()
    
plt.figure()
plt.errorbar(x,sum_prog,yerr=err_prog)
plt.xlabel("steps")
plt.ylabel("T")
plt.grid(True)
plt.title("(Exercise 04.2): Temperature")
plt.show()
```

Trattandosi di uno studio nell'ensemble microcanonico, è possibile osservare come l'energia totale rimanga pressochè costante a differenza della temperatura che oscilla notevolmente.
