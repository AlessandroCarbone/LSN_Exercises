### Distribution probability: $$ p(x)=1 $$

```python
import numpy as np
import math
import matplotlib 
import matplotlib.pyplot as plt

N=100
sum_prog=np.zeros(N)
err_prog=np.zeros(N)
x=np.arange(N)
file=open("output.txt")
line=file.readlines()
for i in range(N):
    data=line[i].split(" ")
    sum_prog[i]=data[0]
    err_prog[i]=data[1]

#Graph 1
plt.errorbar(x,sum_prog,yerr=err_prog)
plt.xlabel("blocks")
plt.ylabel("Integral")
plt.grid(True)
plt.title("(1): Estimation of I with uniform distribution (p(x)=1)")
```

### Distribution probability: $$ p(x)=\frac{\frac{\pi}{2}e^{-\frac{\pi}{2}x}}{1-e^{-\frac{\pi}{2}}} $$

```python
for i in range(N):
    data=line[N+i].split(" ")
    sum_prog[i]=data[0]
    err_prog[i]=data[1]
file.close()

#Graph 2
plt.figure()
plt.errorbar(x,sum_prog,yerr=err_prog)
plt.xlabel("blocks")
plt.ylabel("Integral")
plt.grid(True)
plt.title("(2): Estimation of I with p(x)=$\pi$/2 exp($-\pi$/2 x)/(1-exp(-$\pi$/2)")
plt.show()
```
