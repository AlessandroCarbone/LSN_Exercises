## (1)

```python
import numpy as np
import math
import matplotlib 
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def f(x,k):
    return k*(x**0.5)

N=100
x=np.arange(N)
sum_prog=np.zeros(N)
err_prog=np.zeros(N)

file=open("output.txt")
line=file.readlines()
for i in range(N):
    data=line[i].split(" ")
    sum_prog[i]=data[0]
    err_prog[i]=data[1]
file.close()

plt.errorbar(x,sum_prog,yerr=err_prog,label='simulation')
plt.xlabel("steps")
plt.ylabel("<$|r_N|^2$>")
plt.grid(True)
plt.title("(1): '<$|r_N|^2$> in cubic lattice")

p_opt,_ = curve_fit(f, x, sum_prog)
plt.plot(x,f(x,p_opt[0]),label='fit')
plt.legend()

plt.show()
print("Il parametro k della funzione usata per il fit è:")
print(p_opt[0])
```

## (2)

```python
import numpy as np
import math
import matplotlib 
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def f(x,k):
    return k*(x**0.5)

N=100
x=np.arange(N)
sum_prog=np.zeros(N)
err_prog=np.zeros(N)

file=open("output.txt")
line=file.readlines()
for i in range(N):
    data=line[N+i].split(" ")
    sum_prog[i]=data[0]
    err_prog[i]=data[1]
file.close()

plt.errorbar(x,sum_prog,yerr=err_prog,label='simulation')
plt.xlabel("steps")
plt.ylabel("'<$|r_N|^2$>")
plt.grid(True)
plt.title("(2): '<$|r_N|^2$> in continuum")

p_opt,_ = curve_fit(f, x, sum_prog)
plt.plot(x,f(x,p_opt[0]),label='fit')
plt.legend()

plt.show()
print("Il parametro k della funzione usata per il fit è:")
print(p_opt[0])
```
