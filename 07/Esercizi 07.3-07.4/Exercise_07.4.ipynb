### Osservazioni preliminari

La simulazione seguente parte dalla medesima configurazione iniziale usata per il codice Monte Carlo.

Poichè si tratta di due codici che agiscono in due ensemble differenti occorre simulare le stesse condizioni fisiche per poter avere grandezze confrontabili. In particolare nell'ensemble microcanonico, associato al codice di dinamica molecolare, si ha che la temperatura non è costante, di conseguenza occorre tenere conto dell'equilibrazione del sistema a partire dalle condizioni iniziali per ottenere una temperatura di equilibrio che si avvicina a quella stabilita all'interno del codice MonteCarlo. La convergenza alla temperatura voluta nell'enesemble *NVE* è mostrata opportunamente nei grafici sottostanti.

## $g(r)$

Per valutare la funzione $g(r)$ del gas sono stati utilizzati i dati sugli ultimi $20$ blocchi per tenere conto di una statistica più vicina all'equilibrio. Infatti il gas si distingue dagli altri due stati per la difficoltà a raggiungere un valore di equilibrio, come è possibile vedere dai grafici relativi all'energia interna $U/N$ e la pressione. Per la stessa ragione è stato necessario aumentare il numero di steps della simulazione.


```python
import numpy as np
import math
import matplotlib 
import matplotlib.pyplot as plt

nbins=100
ave_g=np.zeros(nbins)
err_g=np.zeros(nbins)
r=np.zeros(nbins)

ave_g1=np.zeros(nbins)
err_g1=np.zeros(nbins)
r1=np.zeros(nbins)

file1=open("./MonteCarlo_NVT-Results/output.gave.0")
line1=file1.readlines()

for phase in range(1,4):
    if phase==1:
        file=open("gas/ave_gofr.gas")
    if phase==2:
        file=open("liquid/ave_gofr.liquid")
    if phase==3:
        file=open("solid/ave_gofr.solid")
        
    line=file.readlines()
    
    for i in range(nbins):
        data=line[i].split()
        r[i]=data[0]
        ave_g[i]=data[1]
        err_g[i]=data[2]
    
    for j in range(nbins):
        data1=line1[nbins*(phase-1)+j].split()
        r1[j]=data1[0]
        ave_g1[j]=data1[1]
        err_g1[j]=data1[2]
    
    plt.figure()    
    
    if phase==1:
        plt.title("Radial distribution probability $g(r)$ \n Gas phase")
        plt.errorbar(r,ave_g,yerr=err_g,label="MolDyn")
        plt.errorbar(r1,ave_g1,yerr=err_g1,label="MonteCarlo",color='c')
    if phase==2:
        plt.title("Radial distribution probability $g(r)$ \n Liquid phase")
        plt.errorbar(r,ave_g,yerr=err_g,label="MolDyn",color='r')
        plt.errorbar(r1,ave_g1,yerr=err_g1,label="MonteCarlo",color='m')
    if phase==3:
        plt.title("Radial distribution probability $g(r)$ \n Solid phase")
        plt.errorbar(r,ave_g,yerr=err_g,label="MolDyn",color='g')
        plt.errorbar(r1,ave_g1,yerr=err_g1,label="MonteCarlo",color='y')
    
    plt.xlabel("r/$\sigma$")
    plt.ylabel('g(r)')
    plt.legend()
```

Andando ad osservare le immagini sopra riportate è evidente che i due codici restituiscono risultati confrontabili per il valore della $g(r)$. I risultati relativi allo stato gassoso possono essere ulteriormente migliorati con l'ausilio di un numero di dati maggiore.

## Argon : $U/N$, $P$ 

```python
K=1.3806*10**(-23)   #Boltzmann constant

#Data about Argon
sigma=0.34*10**(-9)
rap=120
mass=39.948*1.66054*10**(-27)
eps=rap*K

nblocks=100

x=np.arange(nblocks)
ave_ene=np.zeros(nblocks)
err_ene=np.zeros(nblocks)
ave_ene1=np.zeros(nblocks)
err_ene1=np.zeros(nblocks)
ave_temp=np.zeros(nblocks)
err_temp=np.zeros(nblocks)

ave_press=np.zeros(nblocks)
err_press=np.zeros(nblocks)
ave_press1=np.zeros(nblocks)
err_press1=np.zeros(nblocks)

file3=open("./MonteCarlo_NVT-Results/output.epot.0")
line3=file3.readlines()
file4=open("./MonteCarlo_NVT-Results/output.pres.0")
line4=file4.readlines()

for phase in range(1,4):
    if phase==1:
        file1=open("gas/ave_epot.gas")
        file2=open("gas/ave_press.gas")
        file5=open("gas/ave_temp.gas")
    if phase==2:
        file1=open("liquid/ave_epot.liquid")
        file2=open("liquid/ave_press.liquid")
        file5=open("liquid/ave_temp.liquid")
    if phase==3:
        file1=open("solid/ave_epot.solid")
        file2=open("solid/ave_press.solid")
        file5=open("solid/ave_temp.solid")
    
    line1=file1.readlines()
    line2=file2.readlines()
    line5=file5.readlines()
    
    for i in range(nblocks):
        data1=line1[i].split()
        data2=line2[i].split()
        data5=line5[i].split()
        ave_ene[i]=data1[0]
        err_ene[i]=data1[1]
        ave_press[i]=data2[0]
        err_press[i]=data2[1]
        ave_temp[i]=data5[0]
        err_temp[i]=data5[1]
    
    for j in range(nblocks):
        data3=line3[nbins*(phase-1)+j].split()
        data4=line4[nbins*(phase-1)+j].split()
        ave_ene1[j]=data3[2]
        err_ene1[j]=data3[3]
        ave_press1[j]=data4[2]
        err_press1[j]=data4[3]
    
    fig, (fig_ene, fig_press, fig_temp) = plt.subplots(1, 3, figsize=(18, 4))
    
    if phase==1:
        fig_ene.errorbar(x,ave_ene*eps,yerr=err_ene*eps,label="MolDyn")
        fig_press.errorbar(x,ave_press*eps/(sigma**3),yerr=err_press*eps/(sigma**3),label="MolDyn")
        fig_temp.errorbar(x,ave_temp*rap,yerr=err_temp*rap,label="MolDyn")
        fig_ene.errorbar(x,ave_ene1*eps,yerr=err_ene1*eps,color='c',label="MonteCarlo")
        fig_press.errorbar(x,ave_press1*eps/(sigma**3),yerr=err_press1*eps/(sigma**3),color='c',label="MonteCarlo")
        fig_temp.axhline(y=1.2*rap,color='c',label="MonteCarlo")
        fig_ene.set_title('Internal energy U/N \n Gas phase')
        fig_press.set_title('Pressure P \n Gas phase')
        fig_temp.set_title('Temperature T \n Gas phase')
    if phase==2:
        fig_ene.errorbar(x,ave_ene*eps,yerr=err_ene*eps, color='r',label="MolDyn")
        fig_press.errorbar(x,ave_press*eps/(sigma**3),yerr=err_press*eps/(sigma**3), color='r',label="MolDyn")
        fig_temp.errorbar(x,ave_temp*rap,yerr=err_temp*rap,color='r',label="MolDyn")
        fig_ene.errorbar(x,ave_ene1*eps,yerr=err_ene1*eps,color='m',label="MonteCarlo")
        fig_press.errorbar(x,ave_press1*eps/(sigma**3),yerr=err_press1*eps/(sigma**3),color='m',label="MonteCarlo")
        fig_temp.axhline(y=1.1*rap,color='m',label="MonteCarlo")
        fig_ene.set_title('Internal energy U/N \n Liquid phase')
        fig_press.set_title('Pressure P \n Liquid phase')
        fig_temp.set_title('Temperature T \n Liquid phase')
    if phase==3:
        fig_ene.errorbar(x,ave_ene*eps,yerr=err_ene*eps,color='g',label="MolDyn")
        fig_press.errorbar(x,ave_press*eps/(sigma**3),yerr=err_press*eps/(sigma**3),color='g',label="MolDyn")
        fig_temp.errorbar(x,ave_temp*rap,yerr=err_temp*rap,color='g',label="MolDyn")
        fig_ene.errorbar(x,ave_ene1*eps,yerr=err_ene1*eps,color='y',label="MonteCarlo")
        fig_press.errorbar(x,ave_press1*eps/(sigma**3),yerr=err_press1*eps/(sigma**3),color='y',label="MonteCarlo")
        fig_temp.axhline(y=0.8*rap,color='y',label="MonteCarlo")
        fig_ene.set_title('Internal energy U/N \n Solid phase')
        fig_press.set_title('Pressure P \n Solid phase')
        fig_temp.set_title('Temperature T \n Solid phase')
    
    fig_ene.set_xlabel("blocks")
    fig_ene.set_ylabel("U/N (J)")
    
    fig_press.set_xlabel("blocks")
    fig_press.set_ylabel('P (Pa)')
    
    fig_ene.legend()
    fig_press.legend()
    fig_temp.legend()
```
