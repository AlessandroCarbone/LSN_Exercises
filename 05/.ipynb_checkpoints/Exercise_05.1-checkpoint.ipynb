## Ground state $ \Psi_{1,0,0} $

```python
import numpy as np
import math
import matplotlib 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

N=100
M=1000000
x=np.zeros(M)
y=np.zeros(M)
z=np.zeros(M)
sum_prog=np.zeros(N)
err_prog=np.zeros(N)
steps=np.arange(N)
file1=open("average.txt")
line1=file1.readlines()
for i in range(N):
    data1=line1[i].split(" ")
    sum_prog[i]=data1[0]
    err_prog[i]=data1[1]


#<r> per (1,0,0)
plt.errorbar(steps,sum_prog,yerr=err_prog)
plt.xlabel("steps")
plt.ylabel("<r>")
plt.grid(True)
plt.title("<r> per armonica (1,0,0)")
plt.show()

file2=open("points.txt")
line2=file2.readlines()
for i in range(M):
    data2=line2[i].split(" ")
    x[i]=data2[0]
    y[i]=data2[1]
    z[i]=data2[2]


#Plot dei punti in 3D armonica (1,0,0)
ax = Axes3D(plt.figure())
ax.scatter(x, y, z, c=z, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
plt.show()
```

## 2p excited state $ \Psi_{2,1,0} $

```python
for i in range(N):
    data1=line1[N+i].split(" ")
    sum_prog[i]=data1[0]
    err_prog[i]=data1[1]


#<r> per (2,1,0)
plt.errorbar(steps,sum_prog,yerr=err_prog)
plt.xlabel("steps")
plt.ylabel("<r>")
plt.grid(True)
plt.title("<r> per armonica (2,1,0)")
plt.show()

for i in range(M):
    data2=line2[M+i].split(" ")
    x[i]=data2[0]
    y[i]=data2[1]
    z[i]=data2[2]


#Plot dei punti in 3D armonica (2,1,0)
ax = Axes3D(plt.figure())
ax.scatter(x, y, z, c=z, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
plt.show()
```

## Observations

- We can observe that when we start very far from the origin the convergence of the algorithm is not precise. At the end of the simulation we obtain a value different from the expected one.
- Theoretically speaking, we can say that the ideal width of each Metropolis step gives an acceptance rate of 50%. We have to keep in mind that there is the risk to explore only the nearness of the initial point or to explore places far from the actual position. In both cases we could lose the complete behaviour of the function.
- By using a normal transition probability we see that the outcomes don't differ so much from the outcomes obtained with uniform probability on an interval.
