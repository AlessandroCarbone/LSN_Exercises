## Direct sampling

```python
import numpy as np
import math
import matplotlib 
import matplotlib.pyplot as plt

N=100
sum_prog=np.zeros(N)
err_prog=np.zeros(N)
x=np.arange(N)

file=open("output.txt")
line=file.readlines()

#Call-option
for i in range(N):
    data=line[i].split(" ")
    sum_prog[i]=data[0]
    err_prog[i]=data[1]

plt.figure()
plt.errorbar(x,sum_prog,yerr=err_prog)
plt.xlabel("blocks")
plt.ylabel("C [S(0),0]")
plt.grid(True)
plt.title("Direct sampling: European call-option")
plt.show()
print("Call-option: C[S(0),0]=(",sum_prog[N-1],"+/-",err_prog[N-1],")")

#Put-option
for i in range(N):
    data=line[N+i].split(" ")
    sum_prog[i]=data[0]
    err_prog[i]=data[1]

plt.figure()
plt.errorbar(x,sum_prog,yerr=err_prog)
plt.xlabel("blocks")
plt.ylabel("P [S(0),0]")
plt.grid(True)
plt.title("Direct sampling: European put-option")
plt.show()
print("Put-option: P[S(0),0]=(",sum_prog[N-1],"+/-",err_prog[N-1],")")
```

## Discretized sampling

```python
#Call-option
for i in range(N):
    data=line[2*N+i].split(" ")
    sum_prog[i]=data[0]
    err_prog[i]=data[1]

plt.figure()
plt.errorbar(x,sum_prog,yerr=err_prog)
plt.xlabel("blocks")
plt.ylabel("C [S(0),0]")
plt.grid(True)
plt.title("Discretized sampling: European call-option")
plt.show()
print("Call-option: C[S(0),0]=(",sum_prog[N-1],"+/-",err_prog[N-1],")")

#Put-option
for i in range(N):
    data=line[3*N+i].split(" ")
    sum_prog[i]=data[0]
    err_prog[i]=data[1]

plt.figure()
plt.errorbar(x,sum_prog,yerr=err_prog)
plt.xlabel("blocks")
plt.ylabel("P [S(0),0]")
plt.grid(True)
plt.title("Discretized sampling: European put-option")
plt.show()
print("Put-option: P[S(0),0]=(",sum_prog[N-1],"+/-",err_prog[N-1],")")
```

Dal confronto fra i due tipi di sampling condotti si vede che i risultati sono confrontabili.
