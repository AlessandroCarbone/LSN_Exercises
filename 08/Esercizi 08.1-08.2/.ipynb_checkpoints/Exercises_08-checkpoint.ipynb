## Exercise 08.1

```python
import numpy as np
import math
import matplotlib 
import matplotlib.pyplot as plt

N=100
x=np.arange(N)
sum_prog=np.zeros(N)
err_prog=np.zeros(N)

file=open("ExpectedEigenvalue.txt")
line=file.readlines()
for i in range(N):
    data=line[i].split()
    sum_prog[i]=data[0] 
    err_prog[i]=data[1]
    
plt.errorbar(x,sum_prog,yerr=err_prog)
plt.xlabel("blocks")
plt.ylabel("E")
plt.grid(True)
plt.title("Eigenvalue associated with $\Psi^{\mu,\sigma}$ \n $\mu$=1, $\sigma$=0.5")
plt.show()
```

## Exercise 08.2

### Ground State 

```python
N=100
x=np.arange(N)
sum_prog=np.zeros(N)
err_prog=np.zeros(N)

file=open("GroundState.txt")
line=file.readlines()
for i in range(N):
    data=line[i].split()
    sum_prog[i]=data[0] 
    err_prog[i]=data[1]
    
plt.errorbar(x,sum_prog,yerr=err_prog)
plt.xlabel("blocks")
plt.ylabel("E")
plt.grid(True)
plt.title("Ground State Eigenvalue")
plt.show()
```

### $|\Psi_T (x)|^2$

```python
nbins=600
x=np.zeros(nbins)
pdf=np.zeros(nbins)

file=open("GroundState_pdf.txt")
line=file.readlines()
for i in range(nbins):
    data=line[i].split()
    x[i]=data[0] 
    pdf[i]=data[1]
    
plt.figure(figsize=(8,5))
plt.plot(x,pdf)
plt.xlabel("x")
plt.ylabel("$|\Psi_T (x)|^2$")
plt.grid(True)
plt.title("Ground State pdf")
plt.xlim((-3,3))
plt.show()
```
