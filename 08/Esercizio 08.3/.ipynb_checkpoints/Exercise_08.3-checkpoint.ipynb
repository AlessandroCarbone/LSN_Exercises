## Ground State 

Si osserva che per trovare il ground state si sommano i valori ottenuti di energia cinetica e potenziale. In teoria si può studiare l'errore sull'energia totale usando la propagazione degli errori, ovvero sommando le incertezze di energia cinetica e potenziale. Visto che il nostro interesse è verso l'energia totale, è possibile utilizzare l'errore che è associato alla procedura di data blocking, invece che quello ottenuto dalla propagazione dell'errore.

### $\Psi_T(x)=1$

```python
import numpy as np
import math
import matplotlib 
import matplotlib.pyplot as plt

def error(AV,AV2,n):  # Function for statistical uncertainty estimation
    if n==0:
        return 0
    else:
        return math.sqrt((AV2[n] - AV[n]**2)/n)

N=300
x=np.arange(N)
kin=np.zeros(N)
pot=np.zeros(N)
err_kin=np.zeros(N)
err_pot=np.zeros(N)
ene=np.zeros(N)
err_ene=np.zeros(N)

err_prog=np.zeros(N)
sum_prog=np.zeros(N)
su2_prog=np.zeros(N)

file=open("kinetic_1.dat")
line=file.readlines()
for i in range(N):
    data=line[i].split()
    kin[i]=data[1]
    err_kin[i]=data[2]

file=open("potential_1.dat")
line=file.readlines()
for i in range(N):
    data=line[i].split()
    pot[i]=data[1]
    err_pot[i]=data[2]   #Propagazione dell'errore
file.close()

for i in range(N):
    ene[i]=kin[i]+pot[i]
    err_ene[i]=err_kin[i]+err_pot[i]

#Data blocking
for i in range(N):
    for j in range(i+1):
        sum_prog[i] += ene[j] 
        su2_prog[i] += ene[j]**2 
    sum_prog[i]/=(i+1) # Cumulative average
    su2_prog[i]/=(i+1) # Cumulative square average
    err_prog[i] = error(sum_prog,su2_prog,i) # Statistical uncertainty

plt.errorbar(x,sum_prog,yerr=err_prog)
plt.xlabel("blocks")
plt.ylabel("E")
plt.grid(True)
plt.title("Ground State Eigenvalue \n $\Psi_T(x)=1$")
plt.axhline(-0.46046587969271535,color='r')
plt.show()

print("The progressive sum converge to the energy value: ")
print(sum_prog[N-1])
```

$\Psi_T^{\sigma,\mu}(x) = e^{-\frac{(x-\mu)^2}{2\sigma^2}}+
                               e^{-\frac{(x+\mu)^2}{2\sigma^2}}$

```python
file=open("kinetic_2.dat")
line=file.readlines()
for i in range(N):
    data=line[i].split()
    kin[i]=data[1]
    err_kin[i]=data[2]

file=open("potential_2.dat")
line=file.readlines()
for i in range(N):
    data=line[i].split()
    pot[i]=data[1]
    err_pot[i]=data[2]   #Propagazione dell'errore
file.close()

for i in range(N):
    ene[i]=kin[i]+pot[i]
    err_ene[i]=err_kin[i]+err_pot[i]
    sum_prog[i]=0
    su2_prog[i]=0

#Data blocking
for i in range(N):
    for j in range(i+1):
        sum_prog[i] += ene[j] 
        su2_prog[i] += ene[j]**2 
    sum_prog[i]/=(i+1) # Cumulative average
    su2_prog[i]/=(i+1) # Cumulative square average
    err_prog[i] = error(sum_prog,su2_prog,i) # Statistical uncertainty

plt.errorbar(x,sum_prog,yerr=err_prog)
plt.xlabel("blocks")
plt.ylabel("E")
plt.grid(True)
plt.title("Ground State Eigenvalue \n $\Psi_T^{\sigma,\mu}(x)$")
plt.axhline(-0.46046587969271535,color='r')
plt.show()

print("The progressive sum converge to the energy value: ")
print(sum_prog[N-1])
```

E' possibile vedere come con entrambe le funzioni di prova si ottiene che l'autovalore converge verso il valore che si ottiene analiticamente risolvendo l'equazione agli autovalori. Usando la funzione $\Psi_T^{\mu,\sigma}(x)$ è possibile migliorare la convergenza al valore aspettato. Confrontando i valori finali ottenuti dalla somma progressiva nel primo e nel secondo caso si vede che la differenza tra i due valori è dell'ordine di $10^{-3}$.

## Ground state pdf, $|\Psi_T (x)|^2$


### $\Psi_T(x)=1$ and $ \Psi_T^{\sigma,\mu}(x) = e^{-\frac{(x-\mu)^2}{2\sigma^2}}+
                               e^{-\frac{(x+\mu)^2}{2\sigma^2}}$

```python
from scipy.integrate import simps
import math

def Vpot(x):
    return (x**2 - 2.5)*x**2

def ground_state(x,mu,sigma):
    return math.exp(-(x-mu)**2/(2*sigma**2))+math.exp(-(x+mu)**2/(2*sigma**2))

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

# Step sizes
p = np.linspace(-a/2, a/2, N)
dp = p[1] - p[0] # the step size
V = Vpot(p)

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
H = (-(hbar**2)*CDiff)/(2*m*dp**2) + np.diag(V)
E,psi = np.linalg.eigh(H)
psi = np.transpose(psi)
psi = psi/np.sqrt(dp)

plt.plot(p,(psi[0])**2, color='y', label='GS pdf')

N=300
x=np.zeros(N)
prob=np.zeros(N)
err_prob=np.zeros(N)

file=open("probability_1.dat")
line=file.readlines()
for i in range(N):
    data=line[i].split()
    x[i]=data[0]
    prob[i]=data[1]
    err_prob[i]=data[2]

plt.errorbar(x,prob,yerr=err_prob,label="$\Psi_T(x)=1$")

file=open("probability_2.dat")
line=file.readlines()
for i in range(N):
    data=line[i].split()
    prob[i]=data[1]
    err_prob[i]=data[2]
    
plt.errorbar(x,prob,yerr=err_prob,label="$\Psi_T^{\mu,\sigma}(x)$")

plt.xlabel("x")
plt.ylabel("$|\Psi_T (x)|^2$")
plt.grid(True)
plt.title("Ground State pdf, $|\Psi_T (x)|^2$")
plt.legend()
plt.show()
```

We can see that the ground state probability density function and the values obtain with the two variational wave function are in perfect agreement. However we see that the outcomes don't match exactly with the outcomes of the Variational Monte Carlo code.
