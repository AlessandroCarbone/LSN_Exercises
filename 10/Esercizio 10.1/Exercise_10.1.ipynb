## Circumference

```python
import numpy as np
import math
import matplotlib 
import matplotlib.pyplot as plt
import matplotlib.lines as lines

N=100
sum_prog=np.zeros(N)
err_prog=np.zeros(N)

T=4
temperature=np.array([
    T-i*T/N
    for i in range(N)
])

file=open("Results/circ_output.txt")
line=file.readlines()
for i in range(N):
    data=line[i].split(" ")
    sum_prog[i]=data[1]
    err_prog[i]=data[2]

#Position of the points in the circumference
N_points=30
circ_x=np.zeros(N_points)
circ_y=np.zeros(N_points)

file=open("Results/circ_position.txt")
line=file.readlines()
for i in range(N_points):
    data=line[i].split(" ")
    circ_x[i]=data[0]
    circ_y[i]=data[1]

order=np.zeros(N_points,dtype=int)
for i in range(N_points):
    order[i]=line[N_points+1+i]
file.close()

#(1)
fig_circ, (ax_circ1, ax_circ2) = plt.subplots(1, 2, figsize=(13, 5))

ax_circ1.errorbar(temperature,sum_prog,yerr=err_prog)
ax_circ1.set_xlabel("temperature")
ax_circ1.set_ylabel("<$L^{(1)}$>")
ax_circ1.set_title("Best path <$L^{(1)}$>")
ax_circ1.axhline(y=2*math.pi,color='y',label="$2\pi$")
ax_circ1.legend()

ax_circ2.scatter(circ_x,circ_y)
for i in range(1,N_points):
    ax_circ2.add_line(lines.Line2D([circ_x[order[i-1]-1],circ_x[order[i]-1]],[circ_y[order[i-1]-1],circ_y[order[i]-1]],color='r'))
ax_circ2.add_line(lines.Line2D([circ_x[order[N_points-1]-1],circ_x[order[0]-1]],[circ_y[order[N_points-1]-1],circ_y[order[0]-1]],color='r'))

ax_circ2.set_xlabel("x")
ax_circ2.set_ylabel("y")
ax_circ2.set_title("Circumference path")

```
