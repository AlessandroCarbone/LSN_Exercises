## Equilibration Graphs for T=2

A titolo di esempio si è scelta la temperatura T=2. Il numero di step in ogni blocco è $10^5$ ed è sufficiente per raggiungere l'equilibrazione come si può vedere dai grafici seguenti. Si vede che la variazione delle grandezze fisiche in esame risulta limitata alla terza/quarta cifra decimale.

```python
import numpy as np
import math
import matplotlib 
import matplotlib.pyplot as plt

def error(AV,AV2,n):  # Function for statistical uncertainty estimation
    if n==0:
        return 0
    else:
        return math.sqrt((AV2[n] - AV[n]**2)/n)

N=20
sum_prog=np.zeros(N)
err_prog=np.zeros(N)
steps=np.arange(N)

file1=open("MCIsing_T1/output.ene.1")
line1=file1.readlines()
for i in range(N):
    data1=line1[i].split()
    sum_prog[i]=data1[2]
    err_prog[i]=data1[3] 
file1.close()

#Energy
plt.errorbar(steps,sum_prog,yerr=err_prog)
plt.xlabel("blocks")
plt.ylabel("E")
plt.grid(True)
plt.title("Ising 1D: Internal Energy, T=2")
plt.show()

file2=open("MCIsing_T1/output.heatcapacity.1")
line2=file2.readlines()
for i in range(N):
    data2=line2[i].split()
    sum_prog[i]=data2[2]
    err_prog[i]=data2[3] 
file2.close()  

#Heat Capacity
plt.errorbar(steps,sum_prog,yerr=err_prog)
plt.xlabel("blocks")
plt.ylabel("C")
plt.grid(True)
plt.title("Ising 1D: Heat Capacity, T=2")
plt.show()

file3=open("MCIsing_T1/output.susceptibility.1")
line3=file3.readlines()
for i in range(N):
    data3=line3[i].split()
    sum_prog[i]=data3[2]
    err_prog[i]=data3[3] 
file3.close()  

#Susceptibility
plt.errorbar(steps,sum_prog,yerr=err_prog)
plt.xlabel("blocks")
plt.ylabel("$\chi$")
plt.grid(True)
plt.title("Ising 1D: Susceptibility, T=2")
plt.show()

file4=open("MCIsing_T1/output.magnetization.1")
line4=file4.readlines()
for i in range(N):
    data4=line4[i].split()
    sum_prog[i]=data4[2]
    err_prog[i]=data4[3] 
file4.close()  

#Heat Capacity
plt.errorbar(steps,sum_prog,yerr=err_prog)
plt.xlabel("blocks")
plt.ylabel("M")
plt.grid(True)
plt.title("Ising 1D: Magnetization, T=2")
plt.show()
```

## Temperature Graphs of Internal Energy, Heat Capacity and Susceptibility 

### with J=1, h=0.0

```python
temp=np.zeros(10)
val=np.zeros(10)
err_val=np.zeros(10)

T=2

#Energy
for i in range(10):
    temp[i]=T
    file=open("MCIsing_T"+str(i+1)+"/output.ene.1")
    line=file.readlines()
    data=line[N-1].split()
    val[i]=data[2]
    err_val[i]=data[3]
    T=T-0.15
 
plt.errorbar(temp,val,yerr=err_val,label='Simulation Metropolis',color='r')

for i in range(10):
    file=open("MCIsing_T"+str(i+1)+"/output.ene.0")
    line=file.readlines()
    data=line[N-1].split()
    val[i]=data[2]
    err_val[i]=data[3]
 
plt.errorbar(temp,val,yerr=err_val,label='Simulation Gibbs',color='g')

points=100
#Temperature for the expectation plot
Tt = np.linspace(0.5,2.0,num=points)
beta = 1/Tt
J = 1.0
Ns = 50
th = np.tanh(J/Tt)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )
    
plt.plot(Tt,e,label='Theoretical')

plt.xlabel("T")
plt.ylabel("E")
plt.grid(True)
plt.title("Ising 1D: Internal Energy")
plt.legend()
plt.show()

#Heat Capacity
for i in range(10):
    file=open("MCIsing_T"+str(i+1)+"/output.heatcapacity.1")
    line=file.readlines()
    data=line[N-1].split()
    val[i]=data[2]
    err_val[i]=data[3] 
    
plt.errorbar(temp,val,yerr=err_val,label='Simulation Metropolis',color='r')
    
for i in range(10):
    file=open("MCIsing_T"+str(i+1)+"/output.heatcapacity.0")
    line=file.readlines()
    data=line[N-1].split()
    val[i]=data[2]
    err_val[i]=data[3] 

plt.errorbar(temp,val,yerr=err_val,label='Simulation Gibbs',color='g')
    
heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2) 
plt.plot(Tt, heat,label='Theoretical')

plt.xlabel("T")
plt.ylabel("C")
plt.grid(True)
plt.title("Ising 1D: Heat Capacity")
plt.legend()
plt.show()

#Susceptibility
for i in range(10):
    file=open("MCIsing_T"+str(i+1)+"/output.susceptibility.1")
    line=file.readlines()
    data=line[N-1].split()
    val[i]=data[2]
    err_val[i]=data[3] 
    
plt.errorbar(temp,val,yerr=err_val,label='Simulation Metropolis',color='r')

for i in range(10):
    file=open("MCIsing_T"+str(i+1)+"/output.susceptibility.0")
    line=file.readlines()
    data=line[N-1].split()
    val[i]=data[2]
    err_val[i]=data[3] 
    
plt.errorbar(temp,val,yerr=err_val,label='Simulation Gibbs',color='g')

X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)
plt.plot(Tt,X,label='Theoretical')   

plt.xlabel("T")
plt.ylabel("$\chi$")
plt.grid(True)
plt.title("Ising 1D: Susceptibility")
plt.legend()
plt.show()
```

## Temperature Graph of Magnetization

### with J=1, h=0.02

```python
#Magnetization
for i in range(10):
    file=open("MCIsing_T"+str(i+1)+"/output.magnetization.mag.1")
    line=file.readlines()
    data=line[N-1].split()
    val[i]=data[2]
    err_val[i]=data[3] 
    
plt.errorbar(temp,val,yerr=err_val,label='Simulation Metropolis',color='r')

for i in range(10):
    file=open("MCIsing_T"+str(i+1)+"/output.magnetization.mag.0")
    line=file.readlines()
    data=line[N-1].split()
    val[i]=data[2]
    err_val[i]=data[3] 
    
plt.errorbar(temp,val,yerr=err_val,label='Simulation Gibbs',color='g')

h=0.02 #external field
b = 1/Tt

l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)
plt.plot(Tt, M,label='Theoretical')   

plt.xlabel("T")
plt.ylabel("M")
plt.grid(True)
plt.title("Ising 1D: Magnetization")
plt.legend()
plt.show()

file.close()
```
