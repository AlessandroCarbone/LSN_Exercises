```python
import numpy as np
import math
import matplotlib 
import matplotlib.pyplot as plt

N=10000
s=np.zeros(N)
e=np.zeros(N)
l=np.zeros(N)

file=open("output.txt")
line=file.readlines()
for i in range(10000):
    a=line[i].split(" ")
    s[i]=a[0]
    e[i]=a[1]
    l[i]=a[2]
file.close()

plt.hist(s,50,range=(0,1))

plt.xlabel('Values')
plt.ylabel('Probability')
plt.title('Standard Dice, N=1')
plt.grid(True)

plt.figure()
plt.hist(e,50,color='g',range=(0,7))

plt.xlabel('Values')
plt.ylabel('Probability')
plt.title('Exponential Dice, N=1')
plt.grid(True)

plt.figure()
plt.hist(l,50,color='c',range=(-8,8))

plt.xlabel('Values')
plt.ylabel('Probability')
plt.title('Lorentzian Dice, N=1')
plt.grid(True)

plt.show()
```

```python
import numpy as np
import math
import matplotlib 
import matplotlib.pyplot as plt

N=10000
s=np.zeros(N)
e=np.zeros(N)
l=np.zeros(N)

file=open("output.txt")
line=file.readlines()
for i in range(10000):
    a=line[2*N+i].split(" ")
    s[i]=a[0]
    e[i]=a[1]
    l[i]=a[2]
file.close()

plt.hist(s,50,range=(0.1,0.9))

plt.xlabel('Values')
plt.ylabel('Probability')
plt.title('Standard Dice, N=10')
plt.grid(True)

plt.figure()
plt.hist(e,50,color='g',range=(0,3))

plt.xlabel('Values')
plt.ylabel('Probability')
plt.title('Exponential Dice, N=10')
plt.grid(True)

plt.figure()
plt.hist(l,50,color='c',range=(-8,8))

plt.xlabel('Values')
plt.ylabel('Probability')
plt.title('Lorentzian Dice, N=10')
plt.grid(True)

plt.show()
```

```python
import numpy as np
import math
from math import sqrt,pi
import matplotlib 
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def gauss(x,a,b,c):
    return a*np.exp(-b*(x-c)**2)

def lorentz(x,a,b,c):
    return a/(b**2+(x-c)**2)

N=10000
s=np.zeros(N)
e=np.zeros(N)
l=np.zeros(N)

file=open("output.txt")
line=file.readlines()
for i in range(10000):
    a=line[3*N+i].split(" ")
    s[i]=a[0]
    e[i]=a[1]
    l[i]=a[2]
file.close()

#Grafico Standard
s_heights, s_borders,_ =plt.hist(s,50,range=(0.4,0.6),label='histogram')
s_centers=s_borders[:-1]+np.diff(s_borders)/2

plt.xlabel('Values')
plt.ylabel('Probability')
plt.title('Standard Dice, N=100')
plt.grid(True)

s_int= np.linspace(s_borders[0],s_borders[-1],10000)
s_opt,_= curve_fit(gauss,s_centers,s_heights)
#print(p_opt)
plt.plot(s_int,gauss(s_int,s_opt[0],s_opt[1],s_opt[2]),label='fit',color='r')
plt.legend()

#Grafico Exponential
plt.figure()
e_heights, e_borders,_ =plt.hist(e,50,color='g',range=(0.5,1.5),label='histogram')
e_centers=e_borders[:-1]+np.diff(e_borders)/2

plt.xlabel('Values')
plt.ylabel('Probability')
plt.title('Exponential Dice, N=100')
plt.grid(True)

e_int= np.linspace(e_borders[0],e_borders[-1],10000)
e_opt,_= curve_fit(gauss,e_centers,e_heights)
plt.plot(e_int,gauss(e_int,e_opt[0],e_opt[1],e_opt[2]),label='fit',color='r')
plt.legend()

#Grafico Lorentzian
plt.figure()
l_heights, l_borders,_ =plt.hist(l,50,color='c',range=(-8,8),label='histogram')
l_centers=l_borders[:-1]+np.diff(l_borders)/2

plt.xlabel('Values')
plt.ylabel('Probability')
plt.title('Lorentzian Dice, N=100')
plt.grid(True)

l_int= np.linspace(l_borders[0],l_borders[-1],10000)
l_opt,_= curve_fit(lorentz,l_centers,l_heights)
plt.plot(l_int,lorentz(l_int,l_opt[0],l_opt[1],l_opt[2]),label='fit',color='r')
plt.legend()

plt.show()
```
