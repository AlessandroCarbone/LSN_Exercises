```python
import numpy as np
import math
import matplotlib 
import matplotlib.pyplot as plt

N=100
sum_prog=np.zeros(N)
err_prog=np.zeros(N)
x=np.arange(N)

file=open("output.txt")
line=file.readlines()
for i in range(N):
    data=line[i].split(" ")
    sum_prog[i]=data[0]
    err_prog[i]=data[1]
file.close()

plt.errorbar(x,sum_prog,yerr=err_prog, label='simulation')
plt.axhline(y=math.pi, color='r', label='y=$\pi$')
plt.xlabel("throws")
plt.ylabel("<$\pi$>")
plt.grid(True)
plt.title("Estimation of $\pi$")
plt.legend()
plt.show()
```
