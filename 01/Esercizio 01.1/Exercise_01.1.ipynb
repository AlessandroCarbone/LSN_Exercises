```python
import numpy as np
import math
import matplotlib 
import matplotlib.pyplot as plt

N=100
sum_prog=np.zeros(N)
err_prog=np.zeros(N)
x=np.arange(N)
file=open("output.txt")
line=file.readlines()
for i in range(N):
    a=line[i].split(" ")
    sum_prog[i]=a[0]
    err_prog[i]=a[1]

#(1)
plt.errorbar(x,sum_prog-0.5,yerr=err_prog)
plt.xlabel("throws")
plt.ylabel("<r>-1/2")
plt.grid(True)
plt.title("(1): <r>-1/2")
plt.show()

#(2)
for i in range(N):
    a=line[N+i].split(" ")
    sum_prog[i]=a[0]
    err_prog[i]=a[1]
file.close()

plt.figure()
plt.errorbar(x,sum_prog-1/12,yerr=err_prog)
plt.xlabel("throws")
plt.ylabel("$<r-1/2>^2$-1/12")
plt.grid(True)
plt.title("(2): $<r-1/2>^2$-1/12")
plt.show()

#(3)
for i in range(N):
    a=line[2*N+i].split(" ")
    sum_prog[i]=a[0]
    err_prog[i]=a[1]
file.close()

plt.figure()
plt.errorbar(x,sum_prog,yerr=err_prog)
plt.xlabel("steps")
plt.ylabel("$\chi^2$")
plt.grid(True)
plt.title("(3): $\chi^2$")
plt.show()
print("Il valore di chi-quadro ottenuto  ",sum_prog[N-1]," con errore ",err_prog[N-1])
```
